```python
import pandas as pd
import seaborn as sns
```

```python
salary = pd.read_csv("Salary_dataset.csv")
df = pd.DataFrame(salary)
```

```python
sns.set_theme(style="darkgrid")
regplot = sns.regplot(data=df, x=df['YearsExperience'], y=df['Salary'])
regplot.set_title("Linear Regression: Years of Experience vs Salary",
                  fontdict={
                      "fontsize": 8,
                      "fontweight": 5,
                      },
                  loc="center")
```
